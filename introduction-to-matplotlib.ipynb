```python
hs = pd.read_csv("heart-disease.csv")
hs = hs[hs['age']>50]
hs.head()
```

```python
#pyplot method of scatter plot of age vs cholesterol with target as
#color indicator whether the patient has cancer or not
hs.plot(x='age',y='chol',c='target',kind='scatter');
```

```python
#object oriented method mixed with pyplot 
#scatter plot of age vs cholesterol with target as 
#color plot indicator whether the patient has cancer or not
#see matplotlib plot function's documentation for more customizations
fig,ax = plt.subplots();
hs.plot(x='age',y='chol',c='target',kind='scatter',figsize=(8,8),ax=ax);
```

```python
#object oriented method from scratch
fig,ax = plt.subplots();
scatter = ax.scatter(x=hs['age'],y=hs['chol'],c=hs['target']);

#can also use the method below to only mention col names by using
#data parameter where data = obj name (here hs) of the dataframe/csv 
#scatter = ax.scatter(x='age',y='chol',c='target',data=hs);

#sets the x,y and title labels
ax.set(xlabel='age',ylabel='chol',title="Heart Disease & Cholesterol Levels");

#shows the legend by selecting all legend elements from scatter
ax.legend(*scatter.legend_elements(),title='Target');

#displays the mean across the axis
ax.axhline(hs['chol'].mean(),ls='--');
```

```python
# subplots of age vs chol & heartrate
fig,(ax1,ax2) = plt.subplots(nrows=2,sharex=True,figsize=(10,10))

scatter1 = ax1.scatter(x="age",y="chol",c="target",data=hs)
scatter2 = ax2.scatter(x=hs["age"],y=hs["thalach"],c=hs["target"])

ax1.set(ylabel="chol",title="Cholesterol Levels by Age")
ax2.set(xlabel='age',ylabel='thalach',title="Max Heart Rates by Age")

ax1.legend(*scatter1.legend_elements(),title='Target')
ax2.legend(*scatter2.legend_elements(),title='Target')

ax1.axhline(y=hs["chol"].mean(),ls='--')
ax2.axhline(y=hs['thalach'].mean(),ls='--');

fig.suptitle("Heart Disease",fontsize='16',fontweight='bold');

```

## The Final Plot

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

hs = pd.read_csv('heart-disease.csv')
hs = hs[hs['age']>50]
```

```python
plt.style.use("seaborn-whitegrid")

fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,10))

scatter1 = ax1.scatter(x='age',y='chol',data=hs,c='target',cmap='winter')
scatter2 = ax2.scatter(x='age',y='thalach',data=hs,c='target',cmap='winter')

ax1.set_xlim([50,80])
ax1.set_ylim([100,600])

ax1.set(ylabel='Cholesterol',title='Cholesterol Levels by Age')
ax2.set(xlabel='Age',ylabel='Max Heart Rate',title='Max Heart Rates by Age',ylim=[60,200]);

ax1.legend(*scatter1.legend_elements(),title="Target")
ax2.legend(*scatter2.legend_elements(),title="Target")

ax1.axhline(hs['chol'].mean(),ls='--',color='r',label='Average')
ax2.axhline(hs['thalach'].mean(),ls='--',color='r',label='Average');

fig.suptitle("Heart Disease",fontsize='16',fontweight='bold');
```
